```python
from __future__ import print_function
import os
data_path = ['data']
```

```python
import pandas as pd

# Import the data using the file path
filepath = os.sep.join(data_path + ['heart_attack_prediction_dataset.csv'])
data = pd.read_csv("heart_attack_prediction_dataset.csv")
data
```

```python
data.head(3).T
```

```python
data.drop(['Country', 'Continent', 'Hemisphere','Diet','Sex','Patient ID','Blood Pressure'], axis=1, inplace=True)
data
```

```python
data.columns
```

```python
data.dropna
```

```python
data
```

```python
import warnings
warnings.filterwarnings('ignore', module='sklearn')

from sklearn.preprocessing import MinMaxScaler

msc = MinMaxScaler()

data = pd.DataFrame(msc.fit_transform(data),  # this is an np.array, not a dataframe.
                    columns=data.columns)
```

```python
# Get a list of all the columns that don't contain the label
x_cols = [x for x in data.columns if x != 'Heart Attack Risk']

# Split the data into two dataframes
X_data = data[x_cols]
y_data = data['Heart Attack Risk']
y_data 
X_data
# # alternatively:
# X_data = data.copy()
# y_data = X_data.pop('churned')
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)

knn = knn.fit(X_data, y_data)

y_pred = knn.predict(X_data)
#knn
y_pred
```

```python
def accuracy(real, predict):
    return sum(y_data == y_pred) / float(real.shape[0])
```

```python
print(accuracy(y_data, y_pred))
```

```python
knn = KNeighborsClassifier(n_neighbors=5, weights='distance')

knn = knn.fit(X_data, y_data)

y_pred = knn.predict(X_data)

print(accuracy(y_data, y_pred))
```

```python
knn = KNeighborsClassifier(n_neighbors=5, p=1)

knn = knn.fit(X_data, y_data)

y_pred = knn.predict(X_data)

print(accuracy(y_data, y_pred))
```

```python
score_list = list()

for k in range(1, 21):
    
    knn = KNeighborsClassifier(n_neighbors=k)
    knn = knn.fit(X_data, y_data)
    
    y_pred = knn.predict(X_data)
    score = accuracy(y_data, y_pred)
    
    score_list.append((k, score))
    
score_df = pd.DataFrame(score_list, columns=['k', 'accuracy'])
```

```python
# Import libraries to make the plot

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.set_context('talk')
sns.set_style('ticks')
sns.set_palette('dark')

ax = score_df.set_index('k').plot()

ax.set(xlabel='k', ylabel='accuracy')
ax.set_xticks(range(1, 21));
```
